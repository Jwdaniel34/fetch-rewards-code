```python
Write a program that prints the numbers from 1 to 100.

For every multiple of 3, write Fizz.

For every multiple of 5, write Buzz.

For every multiple of 3 and 5, write FizzBuzz.

For every other number, print out the number.
```

```python
hundred = range(0,100)
for i in hundred:
    if i % 5 == 0 and i % 3 == 0:
        print('FIZZBUZZ',i)
    elif i % 5 == 0:
        print('Buzz', i)
    elif i % 3 == 0:
        print('Fizz', i)    
```
